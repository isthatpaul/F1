```python
import sys
import os

# Add project root to sys.path so src/ can be imported
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import both functions - add load_multiple_seasons
from src.data.load_fastf1 import load_season_results, load_multiple_seasons

# Create cache directory if it doesn't exist
cache_dir = os.path.join(project_root, "cache")
os.makedirs(cache_dir, exist_ok=True)
print(f"✓ Cache directory ready:  {cache_dir}")

# Seaborn style
sns.set_theme(style="whitegrid")

print("✓ Setup complete - ready to load F1 data!")
```

```python
# Load season results for both 2023 and 2024
years = [2023, 2024]

print(f"Loading {years}.. .\n")
season_df = load_multiple_seasons(years, cache_dir="../cache")

print(f"\n{'='*60}")
print("DATA LOADED SUCCESSFULLY")
print(f"{'='*60}")
print(f"Total race results: {len(season_df)}")
print(f"Years: {sorted(season_df['Year'].unique())}")
print(f"Total races: {season_df['Race'].nunique()}")
print(f"Total drivers:  {season_df['DriverFull'].nunique()}")

print(f"\nBreakdown by year:")
for year in years:
    year_data = season_df[season_df['Year'] == year]
    print(f"  {year}:  {year_data['Race'].nunique()} races, {len(year_data)} results")

print(f"{'='*60}")

season_df.head()
```

```python
# Display data info
print("=== Data Info ===")
season_df.info()

print("\n=== Basic Statistics ===")
display(season_df.describe())

print("\n=== Missing Values ===")
missing_values = season_df.isna().sum()
print(missing_values[missing_values > 0])
```

```python
print(f"Total Drivers: {season_df['DriverFull']. nunique()}")
print(f"Drivers:  {sorted(season_df['DriverFull'].unique())}\n")

print(f"Total Races: {season_df['Race'].nunique()}")
print(f"Races: {sorted(season_df['Race'].unique())}")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=season_df, x='Points', bins=20, kde=True)
plt.title("Points Distribution per Race", fontsize=14, fontweight='bold')
plt.xlabel("Points")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Mark DNFs (Position is NaN or > 20)
season_df['DNF'] = season_df['Position']. isna() | (season_df['Position'] > 20)

# Count DNFs
dnf_count = season_df['DNF'].sum()
total_count = len(season_df)
print(f"Total DNFs:  {dnf_count} out of {total_count} races ({dnf_count/total_count*100:.1f}%)")

# Visualize
plt.figure(figsize=(6, 4))
dnf_counts = season_df['DNF']. value_counts()
sns.barplot(x=dnf_counts.index. map({False: 'Finished', True: 'DNF'}), y=dnf_counts.values)
plt.title("DNFs", fontsize=14, fontweight='bold')
plt.ylabel("Count")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
# Mark podiums (Position <= 3)
season_df['Podium'] = (season_df['Position'] <= 3) & (~season_df['Position'].isna())

# Count podiums
podium_count = season_df['Podium'].sum()
print(f"Total Podium Finishes: {podium_count}")

# Visualize
plt. figure(figsize=(6, 4))
podium_counts = season_df['Podium']. value_counts()
sns.barplot(x=podium_counts.index.map({False: 'No Podium', True: 'Podium'}), y=podium_counts.values)
plt.title("Podium Finishes", fontsize=14, fontweight='bold')
plt.ylabel("Count")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
# Group by driver and count podiums
podium_by_driver = season_df[season_df['Podium']].groupby('DriverFull').size().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=podium_by_driver.index, y=podium_by_driver. values, palette='viridis')
plt.title("Podium Finishes by Driver", fontsize=14, fontweight='bold')
plt.xlabel("Driver")
plt.ylabel("Number of Podiums")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate total points per driver
points_by_driver = season_df.groupby('DriverFull')['Points'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=points_by_driver.index, y=points_by_driver.values, palette='coolwarm')
plt.title("Total Points by Driver (2023 & 2024Season)", fontsize=14, fontweight='bold')
plt.xlabel("Driver")
plt.ylabel("Total Points")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nTop 5 Drivers:")
print(points_by_driver.head())
```

```python
# Create output directory if it doesn't exist
output_dir = "data/processed"
os.makedirs(output_dir, exist_ok=True)

# Save to CSV
output_path = os.path.join(output_dir, "season_2023_raw.csv")
season_df.to_csv(output_path, index=False)
print(f"✓ Season data saved to {output_path}")
print(f"  Shape: {season_df.shape}")
print(f"  Columns: {list(season_df.columns)}")
```
